```python
def calculate_module_of_arrays(our_fields: np.ndarray) -> np.ndarray:
    module_of_fields = np.zeros((our_fields[0].shape))
    module_of_fields = (our_fields[0]**2 + our_fields[1]**2 + our_fields[2]**2)**0.5
    return module_of_fields
```

```python
data = StreamData("myfile_3d.csv", 501, 981, 81)
resultY_start = advance_to_column(data, "Y Velocity")
resultX_start = advance_to_column(data, "X Velocity")
resultZ_start = advance_to_column(data, "Z Velocity")
module_start = calculate_module_of_arrays([resultX_start, resultY_start, resultZ_start])
data = StreamData("gauss_iter_3.out.csv", 501, 981, 81)
resultY_start = advance_to_column(data, "Y Velocity")
resultX_start = advance_to_column(data, "X Velocity")
resultZ_start = advance_to_column(data, "Z Velocity")
module_gauss_3 = calculate_module_of_arrays([resultX_start, resultY_start, resultZ_start])
data = StreamData("gauss_iter_6.out.csv", 501, 981, 81)
resultY_start = advance_to_column(data, "Y Velocity")
resultX_start = advance_to_column(data, "X Velocity")
resultZ_start = advance_to_column(data, "Z Velocity")
module_gauss_6 = calculate_module_of_arrays([resultX_start, resultY_start, resultZ_start])
```
