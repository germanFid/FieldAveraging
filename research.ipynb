```python
import averager
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
from matplotlib.ticker import LinearLocator


def research(inputed_field):
    our_figure = plt.figure()
    our_plot = our_figure.add_subplot(111, projection='3d')
    n = len(inputed_field[0])
    m = len(inputed_field)
    x = range(n)
    y = range(m)
    X, Y = np.meshgrid(x, y) 
    surf = our_plot.plot_surface(X, Y, inputed_field, cmap='RdYlGn', linewidth=0, antialiased=False)
    our_plot.zaxis.set_major_locator(LinearLocator(5))
    our_plot.zaxis.set_major_formatter('{x:.02f}')
    our_figure.colorbar(surf, shrink=0.5, aspect=10)

    plt.show()


def redefinition_array(inputed_field):
    n = len(inputed_field[0])
    m = len(inputed_field)
    outputed_field = np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            outputed_field[i][j] = inputed_field[i][j]
    return outputed_field


def test():
    averaging_width = 1
    w, h = 5, 3
    input_field=np.zeros((w,h))
    my_array=np.zeros((w,h))
    # input_field = [[float(0) for x in range(w)] for y in range(h)]
    input_field[4][2] = 15
    input_field[2][1] = 15
    research(input_field)
    # print(line_2_row_averaging_this(input_field, averaging_width))
    my_array = averager.diagonally_averaging_this(input_field, averaging_width)
    my_array = redefinition_array(my_array)
    research(my_array)
    for i in range(4):
        my_array = averager.diagonally_averaging_this(my_array, averaging_width)
        my_array = redefinition_array(my_array)
        research(my_array)
    

test()

```
