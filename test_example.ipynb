```python
import matplotlib.pyplot as plt
import usefull_graphics as ug
import averager as av
import numpy as np
```

```python
# Set the size of the figure
figsize = (16, 8)

# Remove the axes from the subplots
show_axes = False

# Add a color bar to the subplots
show_colorbar = True

# Set the width ratio of each subplot
subplot_widths = [4, 4, 4, 4]

# Set the font size for all text elements
font_size = 10

titles = ['test']

a = 30
norm = [-a, a]

COMMON_RADIUS = 4
COMMON_PROCESSES = 8
```

```python
arr = np.zeros((50, 50))
shape = arr[14:34, 14:34].shape
x, y = np.meshgrid(range(arr.shape[0]), range(arr.shape[1]))
arr[14:34, 14:34] = 30
arr1 = arr 
arr2 = arr
arr1 = av.average_2d_by_gauss(arr1, 3*COMMON_RADIUS)
for i in range(5):
    arr2 = av.basic_2d_array_averaging_parallel(arr2, COMMON_RADIUS, COMMON_PROCESSES, visuals=True)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
ax1.plot_surface(x, y, arr)
plt.title('original, sum = ' + str(np.sum(arr)))
plt.show()

fig2 = plt.figure()
ax3 = fig2.add_subplot(111, projection='3d')
ax3.plot_surface(x, y, arr2)
plt.title('basic, iter = 5, sum = ' + str(np.sum(arr2)))
plt.show()

fig1 = plt.figure()
ax2 = fig1.add_subplot(111, projection='3d')
ax2.plot_surface(x, y, arr1)
plt.title('gauss, sum = ' + str(np.sum(arr1)))
plt.show()
```

```python
fig.savefig("test.png", dpi = 300, transparent=True)
fig1.savefig("test1.png", dpi = 300, transparent=True)
fig2.savefig("test2.png", dpi = 300, transparent=True)
```
